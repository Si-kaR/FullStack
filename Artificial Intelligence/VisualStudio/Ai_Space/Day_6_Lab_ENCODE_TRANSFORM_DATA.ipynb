```python
from sklearn.impute import SimpleImputer
import pandas as pd

import numpy as np
```

```python
from numpy import array
from sklearn.preprocessing import LabelEncoder
# define example
data = ['warm','hot','cold','luke warm','hot','cold','warm']
```

```python
values = array(data)
print(values)
```

```python
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(values)
print(integer_encoded)
```

```python
inverted = label_encoder.inverse_transform(integer_encoded)[4]
print(inverted)
```

```python
covid_result=['Positive','Negative','Positive','Negative','Negative',]
```

```python
# Now attemtping to Encode
# Display should be series
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(covid_result)


print(integer_encoded)
```

So our above is an array.
Change it to a series



```python
integer_to_series = integer_encoded

print(integer_to_series)
```

```python
series = pd.Series(integer_to_series)
print(series)
```
